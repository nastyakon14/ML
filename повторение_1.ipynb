5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
matr_1 = np.zeros(25)
matr_1[::2] = 5
A = matr_1.reshape(5,5); A
```

```python
from numpy.linalg import*
```

```python
np.linalg.det(A)
```

```python
def inv(matr):
  try:
    assert np.linalg.det(matr) != 0, 'Определитель матрицы не ноль'
    return (np.linalg.inv(matr))
  except AssertionError:
    return('Определитель = 0 => обратной матрицы не существует')

inv(A)
```

```python
matr_2 = np.array([5]*25).reshape(5,5)
B = np.triu(matr_2); B
```

```python
print(np.linalg.det(B))
print(np.linalg.inv(B))
```

```python
inv(B)
```

19.  диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,10))
fig.suptitle('Гистограммы рассеивания', size = 20, color = 'blue')
ax[0,0].set_xlabel('Price')
ax[0,1].set_xlabel('Price')
ax[1,0].set_xlabel('Price')
ax[1,1].set_xlabel('Price')
ax[0,0].set_ylabel('Rooms')
ax[0,1].set_ylabel('Square')
ax[1,0].set_ylabel('HouseFloor')
ax[1,1].set_ylabel('HouseYear')
ax[0,0].set_title('Гистогамма рассевания для признаков Rooms', size = 10)
ax[0,1].set_title('Гистогамма рассевания для признаков Square', size = 10)
ax[1,0].set_title('Гистогамма рассевания для признаков HouseFloor', size = 10)
ax[1,1].set_title('Гистогамма рассевания для признаков HouseYear', size = 10)

ax[0,0].grid(), ax[0,1].grid(), ax[1,0].grid(), ax[1,1].grid()

sc1 = ax[0,0].scatter(data.Price, data.Rooms, color ='orange')
sc2 = ax[0,1].scatter(data.Price, data.Square, color = 'green')
sc3 = ax[1,0].scatter(data.Price, data.HouseFloor)
sc4 = ax[1,1].scatter(data.Price, data.HouseYear, color = 'yellow')

plt.legend((sc1, sc2, sc3, sc4),
           ('Rooms', 'Square', 'HouseFloor', 'HouseYear'),
           ncol=1,    # расположение в 1 столбец
           fontsize=15,   # размер шрифта
           bbox_to_anchor=(1, 0, 0.5, 1.25))



plt.show()
```
